THEORETICAL ANSWERS

1. What is NumPy, and why is it widely used in Python?
* NumPy is a core library for handling numerical data in Python. It offers tools to work efficiently with arrays and perform complex mathematical operations. It’s preferred because it boosts performance and integrates well with other data libraries.

2. How does broadcasting work in NumPy?
* Broadcasting allows operations between arrays of different shapes by automatically aligning dimensions. It reduces the need for manual reshaping and helps perform element-wise operations without writing loops.

3. What is a Pandas DataFrame?
* A DataFrame is a tabular data structure in Pandas. It looks like a spreadsheet and allows easy handling, filtering, and analysis of data with rows and columns.

4. Explain the use of the groupby() method in Pandas.
* The groupby() function splits data into groups based on a column’s value, then allows applying functions like sum, mean, or count to each group. It's commonly used for summarizing data.

5. Why is Seaborn preferred for statistical visualizations?
* Seaborn is a plotting library that makes it easy to create graphs with statistical meaning. It works well with Pandas and simplifies the process of building clean and informative visualizations.

6. What are the differences between NumPy arrays and Python lists?
* NumPy arrays are more efficient in terms of speed and memory. They support mathematical operations directly, unlike Python lists, which are slower and lack such functionality.

7. What is a heatmap, and when should it be used?
* A heatmap shows data using colors to represent values. It’s especially useful for identifying trends, correlations, or high/low values in a matrix.

8. What does “vectorized operation” mean in NumPy?
* Vectorized operations allow performing calculations on entire arrays at once, without writing explicit loops. This approach improves speed and code readability.

9. How does Matplotlib differ from Plotly?
* Matplotlib creates static plots suited for reports and publications, while Plotly provides interactive, web-ready charts with advanced features like zooming and hovering.

10. What is the significance of hierarchical indexing in Pandas?
* Hierarchical indexing allows multi-level row or column labels. It helps organize complex datasets and makes it easier to group and retrieve data in multi-dimensional formats.

11. What is the role of Seaborn’s pairplot() function?
* pairplot() draws scatter plots for each pair of features in a dataset, giving a visual overview of relationships and distributions in one step.

12. What is the purpose of the describe() function in Pandas?
* The describe() method provides a summary of numerical columns, showing stats like mean, standard deviation, and quartiles to understand the dataset quickly.

13. Why is handling missing data important in Pandas?
* Missing data can lead to incorrect conclusions. Handling it ensures the analysis is accurate and the data is complete or adjusted properly before processing.

14. What are the benefits of using Plotly for data visualization?
* Plotly creates interactive and visually appealing graphs. It’s great for dashboards and allows users to explore data by interacting with the plots.

15. How does NumPy handle multidimensional arrays?
* NumPy supports arrays with multiple dimensions and provides tools to manipulate, slice, and reshape them easily, which is helpful for advanced mathematical computations.

16. What is the role of Bokeh in data visualization?
* Bokeh is used to create interactive plots that can be embedded into web applications. It’s useful for visualizing large datasets in real-time.

17. Explain the difference between apply() and map() in Pandas.
* map() is used for simple element-wise transformations on a Series. apply() is more flexible and can apply functions to rows or columns of a DataFrame.

18. What are some advanced features of NumPy?
* Advanced NumPy features include broadcasting, random number generation, linear algebra operations, masking, and memory-mapped arrays for large datasets.

19. How does Pandas simplify time series analysis?
* Pandas has built-in support for datetime objects and functions to resample, shift, and filter data based on time, making time-based analysis much easier.

20. What is the role of a pivot table in Pandas?
* A pivot table rearranges and summarizes data based on one or more keys, allowing you to view aggregate values like totals or averages in a structured format.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
* NumPy arrays are stored in a contiguous memory block, making slicing faster. Python lists are more flexible but involve more overhead in memory and execution.

22. What are some common use cases for Seaborn?
* Seaborn is often used to create plots that show distributions, relationships, and trends — like histograms, heatmaps, box plots, and line graphs.*

1. How do you create a 2D NumPy array and calculate the sum of each row?

```python
import numpy as np

# Creating a 2D NumPy array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Calculating the sum of each row
row_sum = np.sum(arr, axis=1)

print("Row-wise sum:", row_sum)
```

2. Pandas mean of a column

```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Marks': [85, 90, 78]}
df = pd.DataFrame(data)

# Mean of 'Marks' column
mean_marks = df['Marks'].mean()

print("Average Marks:", mean_marks)
```

3. Scatter plot using Matplotlib

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [15, 25, 35, 45]

plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

4. Seaborn correlation matrix with heatmap

```python
import seaborn as sns
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'Math': [90, 85, 70, 60],
    'Science': [88, 82, 75, 65],
    'English': [80, 78, 72, 68]
})

# Correlation matrix
corr_matrix = df.corr()

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

5. Bar plot using Plotly

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame({'Fruits': ['Apple', 'Banana', 'Orange'], 'Quantity': [10, 15, 7]})
fig = px.bar(data, x='Fruits', y='Quantity', title='Fruit Quantities')
fig.show()
```

6. Add column based on another in Pandas

```python
import pandas as pd

df = pd.DataFrame({'Product': ['A', 'B', 'C'], 'Price': [100, 200, 300]})
df['Discounted_Price'] = df['Price'] * 0.9

print(df)
```

7. Element-wise multiplication (NumPy)

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print("Element-wise multiplication:", result)
```

8. Multiple line plots using Matplotlib

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2', linestyle='--')
plt.title("Line Plot with Two Lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

9. Filter DataFrame rows by condition

```python
import pandas as pd

df = pd.DataFrame({'Name': ['Yav', 'Ansh', 'Hope'], 'Score': [65, 85, 75]})
filtered_df = df[df['Score'] > 70]

print(filtered_df)
```

10. Histogram using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [60, 65, 70, 70, 75, 80, 85, 90, 95, 100]

sns.histplot(data, bins=5, kde=True)
plt.title("Score Distribution")
plt.show()
```

11. Matrix multiplication using NumPy

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

result = np.dot(A, B)
print("Matrix Multiplication Result:\n", result)
```

13. 3D scatter plot using Plotly

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
